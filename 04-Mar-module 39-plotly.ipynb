Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.


```python
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Create a scatter plot for age and fare columns
fig = px.scatter(titanic, x="age", y="fare", title="Scatter plot of Age vs Fare in Titanic Dataset")

fig.show()
```

Q3. Using the tips dataset in the Plotly library,
 Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset.
 Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px

# Load the tips dataset from Plotly
tips = px.data.tips()

# Create a histogram using Plotly Express
fig = px.histogram(
    tips, 
    x="sex", 
    y="total_bill", 
    color="day", 
    pattern_shape="smoker",
    title="Histogram of Total Bill by Sex with Smoker and Day Details"
)

fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.


```python
import plotly.express as px

# Load the iris dataset from Plotly
iris = px.data.iris()

# Create a scatter matrix plot using Plotly Express
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset"
)

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

A distplot (distribution plot) is a visualization that shows the distribution of a dataset. It can display the data distribution as a histogram, along with a fitted density plot, and other related plots. Distplots are useful for understanding the distribution, central tendency, and variability of data.

Plotly Express doesn't have a specific distplot function like Seaborn. However, you can create a similar visualization using a combination of histogram and density plots.

```python
import plotly.express as px
import plotly.figure_factory as ff

# Load the tips dataset from Plotly
tips = px.data.tips()

# Create a distplot using Plotly Figure Factory
fig = ff.create_distplot(
    [tips['total_bill']], 
    group_labels=['total_bill'], 
    bin_size=2
)

fig.update_layout(title="Distplot of Total Bill")

fig.show()
```
