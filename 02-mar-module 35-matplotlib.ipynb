Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for its ability to produce publication-quality figures and integrate seamlessly with many GUI toolkits, web application servers, and frameworks. 

### Why is Matplotlib Used?
1. **Versatility**: It can create a wide range of plots and visualizations.
2. **Integration**: Works well with NumPy, pandas, and other scientific libraries.
3. **Customization**: Highly customizable to meet specific requirements.
4. **Interactivity**: Supports interactive plots and can embed plots in applications.
5. **Compatibility**: Compatible with various backends, enabling use in different environments (e.g., Jupyter notebooks, web applications).

### Five Plots Using the Pyplot Module of Matplotlib

1. **Line Plot**
   ```python
   import matplotlib.pyplot as plt
   plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
   plt.title('Line Plot')
   plt.xlabel('X-axis')
   plt.ylabel('Y-axis')
   plt.show()
   ```

2. **Scatter Plot**
   ```python
   import matplotlib.pyplot as plt
   plt.scatter([1, 2, 3, 4], [10, 20, 25, 30])
   plt.title('Scatter Plot')
   plt.xlabel('X-axis')
   plt.ylabel('Y-axis')
   plt.show()
   ```

3. **Bar Plot**
   ```python
   import matplotlib.pyplot as plt
   plt.bar(['A', 'B', 'C', 'D'], [3, 7, 8, 5])
   plt.title('Bar Plot')
   plt.xlabel('Categories')
   plt.ylabel('Values')
   plt.show()
   ```

4. **Histogram**
   ```python
   import matplotlib.pyplot as plt
   data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
   plt.hist(data, bins=4)
   plt.title('Histogram')
   plt.xlabel('Value')
   plt.ylabel('Frequency')
   plt.show()
   ```

5. **Pie Chart**
   ```python
   import matplotlib.pyplot as plt
   labels = 'A', 'B', 'C', 'D'
   sizes = [15, 30, 45, 10]
   plt.pie(sizes, labels=labels, autopct='%1.1f%%')
   plt.title('Pie Chart')
   plt.show()
   ```

These examples illustrate the basic usage of Matplotlib's Pyplot module to create common types of plots. Each plot type can be customized extensively to suit different needs.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.


A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. If the points are color-coded, one additional variable can be displayed. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.



````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
````

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
 

The subplot() function in Matplotlib is used to create multiple plots in a single figure. This is particularly useful when comparing multiple datasets or visualizations side by side. The function divides the figure into a grid and allows you to place different plots in each cell of the grid.

Drawing Four Line Plots Using the subplot() Function :

````python
import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x1, y1 = np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])
x2, y2 = np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])
x3, y3 = np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])
x4, y4 = np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150])

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot for line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Plot for line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

# Plot for line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

# Plot for line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
````

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
 

What is a Bar Plot?
A bar plot is a chart that presents categorical data with rectangular bars. Each bar's length is proportional to the value it represents. Bar plots are used to compare different categories or groups, making it easy to visualize the differences in value between them.

Why is a Bar Plot Used?

1.Comparison: Helps in comparing different categories or groups.

2.Clarity: Provides a clear and straightforward way to present data.

3.Versatility: Can display both positive and negative values.

4.Visualization: Makes it easier to understand the relative sizes of different data points

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(10, 5))
plt.bar(company, profit, color='skyblue')
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit in USD')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(10, 5))
plt.barh(company, profit, color='skyblue')
plt.title('Company Profits')
plt.xlabel('Profit in USD')
plt.ylabel('Company')
plt.show()
````

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It may also include "whiskers" that extend to show the range of the data and any outliers.

Why is a Box Plot Used?

-Summary Statistics: Provides a visual summary of key statistics of the data.

-Comparison: Helps in comparing the distributions between different datasets.

-Outliers Detection: Identifies potential outliers within the data.

-Distribution Shape: Shows the skewness and spread of the data

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Two Distributions')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.show()
````
