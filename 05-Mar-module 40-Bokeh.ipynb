Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example

In Bokeh, glyphs are the basic visual building blocks of a plot. They represent the shapes drawn on the plot, such as circles, lines, rectangles, etc. Each glyph is associated with data that specifies its position, size, color, and other visual properties. You can add glyphs to a Bokeh plot to visualize your data in various forms.

Here’s how you can add different types of glyphs to a Bokeh plot:

-Import Bokeh libraries: Import the necessary modules from Bokeh.

-Prepare your data: You can use lists, NumPy arrays, or Pandas DataFrames for your data.

-Create a figure: Use the figure function to create a new plot.

-Add glyphs: Use functions like circle, line, square, etc., to add glyphs to your plot.

-Show or save the plot: Use the show function to display the plot or the save function to save it.

example :

```python
# Step 1: Import Bokeh libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 5: Display plots in the Jupyter notebook
output_notebook()

# Step 2: Prepare your data
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [7, 6, 3, 8, 2]
y3 = [4, 5, 1, 2, 3]

# Step 3: Create a figure
p = figure(title="Multiple Glyphs Example", x_axis_label='X-Axis', y_axis_label='Y-Axis')

# Step 4: Add different types of glyphs
# Add a line glyph
p.line(x, y1, legend_label="Line", line_width=2, color="blue")

# Add circle glyphs
p.circle(x, y2, size=10, legend_label="Circles", color="red", alpha=0.5)

# Add square glyphs
p.square(x, y3, size=10, legend_label="Squares", color="green", alpha=0.5)

# Step 5: Show the plot
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


Customizing the appearance of a Bokeh plot involves several aspects, including the plot's title, axes, legend, and overall styling. Below are some examples of how to customize each of these components:

1. Customizing the Plot Title
You can customize the plot title by setting its properties like text_font_size, text_color, align, etc.

2. Customizing Axes
Axes customization can include setting labels, ticks, tick labels, and their appearances such as font size, color, orientation, etc.

3. Customizing the Legend
Legends can be customized by setting their location, orientation, background color, border color, label properties, and more.

Example
Here’s a comprehensive example demonstrating how to customize the appearance of a Bokeh plot:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Title

# Display plots in the Jupyter notebook
output_notebook()

# Prepare your data
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [7, 6, 3, 8, 2]
y3 = [4, 5, 1, 2, 3]

# Create a figure
p = figure(title="Customized Bokeh Plot", x_axis_label='X-Axis', y_axis_label='Y-Axis')

# Add a line glyph
p.line(x, y1, legend_label="Line", line_width=2, color="blue")

# Add circle glyphs
p.circle(x, y2, size=10, legend_label="Circles", color="red", alpha=0.5)

# Add square glyphs
p.square(x, y3, size=10, legend_label="Squares", color="green", alpha=0.5)

# Customize plot title
p.title.text_font_size = '20pt'
p.title.text_color = 'navy'
p.title.align = 'center'
p.add_layout(Title(text="Subtitle: An example of customized Bokeh plot", text_font_size="12pt", text_color="gray"), 'above')

# Customize axes
p.xaxis.axis_label_text_font_size = '14pt'
p.xaxis.axis_label_text_color = 'darkred'
p.xaxis.major_label_text_font_size = '12pt'
p.xaxis.major_label_text_color = 'darkblue'

p.yaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_color = 'darkred'
p.yaxis.major_label_text_font_size = '12pt'
p.yaxis.major_label_text_color = 'darkblue'

# Customize legend
p.legend.title = 'Legend'
p.legend.title_text_font_size = '14pt'
p.legend.title_text_color = 'black'
p.legend.label_text_font_size = '12pt'
p.legend.label_text_color = 'black'
p.legend.location = 'top_left'
p.legend.background_fill_color = 'lightgray'
p.legend.border_line_color = 'black'

# Customize grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
p.xgrid.grid_line_dash = [6, 4]
p.ygrid.grid_line_dash = [6, 4]

# Show the plot
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A Bokeh server is a Python server application that allows you to create interactive, data-driven plots and dashboards that can be updated in real time. It is particularly useful for applications where the data changes dynamically, such as live data feeds, user interactions, or complex simulations.

Key Features of Bokeh Server:

1.Real-time Updates: Automatically updates plots when the underlying data changes.

2.Interactive Widgets: Integrates with widgets like sliders, dropdowns, buttons, etc., to allow user interaction.

3.Streaming Data: Supports streaming data to continuously update plots


##Basic Steps to Create a Bokeh Server Application:

Install Bokeh: Make sure Bokeh is installed.

Import Necessary Libraries: Import Bokeh libraries for plotting and server interaction.

Define the Plot and Widgets: Create plots and widgets.

Create Update Functions: Define functions to update plots based on user interactions or real-time data.

Define the Layout: Arrange plots and widgets in a layout.

Run the Bokeh Server: Use the bokeh serve command to run the application.
