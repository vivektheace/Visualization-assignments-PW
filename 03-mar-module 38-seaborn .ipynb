Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Sure, here are five plots that can be created using the Seaborn library, along with their uses:

1. **Scatter Plot (`sns.scatterplot`)**
   - **Use**: Scatter plots are used to examine the relationship between two continuous variables. They can help identify correlations, clusters, and outliers in the data.

2. **Box Plot (`sns.boxplot`)**
   - **Use**: Box plots are used to show the distribution of a dataset based on a five-number summary: minimum, first quartile, median, third quartile, and maximum. They are useful for comparing distributions and identifying outliers.

3. **Histogram (`sns.histplot`)**
   - **Use**: Histograms are used to visualize the distribution of a single continuous variable. They show the frequency of data points within specified ranges (bins) and help understand the underlying frequency distribution.

4. **Heatmap (`sns.heatmap`)**
   - **Use**: Heatmaps are used to represent data in a matrix format where individual values are represented as colors. They are particularly useful for visualizing the correlation matrix of a dataset or for showing the intensity of data at various intersections.

5. **Bar Plot (`sns.barplot`)**
   - **Use**: Bar plots are used to compare the means or sums of different categories. They are useful for displaying categorical data with rectangular bars, where the length of each bar is proportional to the value it represents.

These plots provide powerful ways to visualize and analyze data, making it easier to understand patterns, relationships, and distributions.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.Â 

Note: timepoint, signal, event, and region are columns in the fmri dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)

plt.title('FMRI Signal Over Time by Event and Region')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event and Region')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.


```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Plot box plots for 'age' and 'fare' against 'pclass'
plt.figure(figsize=(14, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
# Plot a histogram for the 'price' column with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')

plt.title('Distribution of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
# Pivot the dataset to create a matrix format suitable for a heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.title('Heatmap of Monthly Passengers Over Years')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
